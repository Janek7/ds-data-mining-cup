```python
import pandas as pd
import random
```

```python
items=pd.read_csv("items.csv", sep="|")
items.head()
```

```python
transactions=pd.read_csv("transactions.csv", sep="|")
transactions.head()
```

```python
evaluation=pd.read_csv("evaluation.csv")
evaluation
```

```python
#recommend books of the same author
def recommendation(book_id):
    #default recommendations with random books

    df=pd.DataFrame({"itemID": [book_id], "rec_1": [items.iloc[random.randrange(len(items)),0]], "rec_2": [items.iloc[random.randrange(len(items)),0]], "rec_3":[items.iloc[random.randrange(len(items)),0]], "rec_4":[items.iloc[random.randrange(len(items)),0]], "rec_5":[items.iloc[random.randrange(len(items)),0]]})
    
    #update default recommendations with book of the same author if posiible
        
    if not items[items["itemID"]==book_id].empty:
        author=items[items["itemID"]==book_id]["author"] # save the author
        books=items[items["author"]==author.iloc[0]] #filter the books pf the same author
        books=books[books["itemID"]!=book_id] #drop the book for which we look up recommendations

            #check the size
        if len(books)==5:
                #fill in the recommendations
                df=pd.DataFrame({"itemID": [book_id], "rec_1": [books.iloc[0,0]], "rec_2": [books.iloc[1,0]], "rec_3":[books.iloc[2,0]], "rec_4":[books.iloc[3,0]], "rec_5":[books.iloc[4,0]]})
        if len(books)>5:
                #randomly choose 5 books and fill in the recommendations
                my_list=random.sample(books["itemID"].tolist(),5)
                df["rec_1"]=my_list[0]
                df["rec_2"]=my_list[1]
                df["rec_3"]=my_list[2]
                df["rec_4"]=my_list[3]
                df["rec_5"]=my_list[4]
        if len(books)==4:
                df["rec_1"]=books.iloc[0,0]
                df["rec_2"]=books.iloc[1,0]
                df["rec_3"]=books.iloc[2,0]
                df["rec_4"]=books.iloc[3,0]
        if len(books)==3:
                df["rec_1"]=books.iloc[0,0]
                df["rec_2"]=books.iloc[1,0]
                df["rec_3"]=books.iloc[2,0]
        if len(books)==2:
                df["rec_1"]=books.iloc[0,0]
                df["rec_2"]=books.iloc[1,0]
        if len(books)==1:
                df["rec_1"]=books.iloc[0,0]

    print(df)
    
    return df
```

```python
#Compiling recommendation matrix out of individual recs
full_recommendation=pd.DataFrame(columns=["itemID","rec_1","rec_2", "rec_3","rec_4","rec_5"])
for i in range(len(evaluation)):
    r=recommendation(evaluation.iloc[i].iloc[0])
    full_recommendation=full_recommendation.append(r, ignore_index=True)
```
